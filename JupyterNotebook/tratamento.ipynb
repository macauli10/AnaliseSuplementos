Renomeando as colunas e tirando os espaços

```python
df.columns
```

```python
traduzindo = {
    'Date': 'data',
    'Product Name': 'nome_produto',
    'Category': 'categoria',
    'Units Sold': 'unidades_vendidas',
    'Price': 'preco',
    'Revenue': 'receita',
    'Discount': 'desconto',
    'Units Returned': 'unidades_devolvidas',
    'Location': 'localizacao',
    'Platform': 'plataforma'
}

df = df.rename(columns=traduzindo)
```

Tranformando em banco de dados SQL para fazer consultas

```python
conn = sqlite3.connect("dados_suplementos.db")


df.to_sql("tabela_suplementos", conn, if_exists="replace", index=False)

conn.close()
```

```python
df.to_csv('dados_tratados.csv', index=False)
```
